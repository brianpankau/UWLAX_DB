# Problem 1.  Revenue models.

Average monthly revenue at each store in the TableFarm salad chain is $\$100,000$, with a standard deviation of $\$12,000$.  An advertising firm claims they can increase monthly revenue to $\$120,000$, but the standard deviation will be increased as well, to $\$25,000$.

Write Python code to generate three lists of random numbers which model potential revenue: one list with 12 months of revenue using the current mean and standard deviation, another list with 12 months of revenue using the predicted mean and standard deviation, and a third list combining your first two lists.  You can assume a normal distribution.  Round each number to the nearest $\$1,000$.

````python
# AUTHOR: Brian Pankau
# DATE: 16 June 2017
# CLASS: UWLAX DS710 SUMMER 2017 
# ASSIGNMENT: 4-1 Python program 
# EXCUTE: python 
#
# Problem 1: Revenue Models
# Average monthly revenue at each store in the TableFarm salad chain 
# is $100,000, with a standard deviation of $12,000. An advertising 
# firm claims they can increase monthly revenue to $120,000, but the 
# standard deviation will be increased as well, to $25,000.
# Write Python code to generate three lists of random numbers which 
# model potential revenue: one list with 12 months of revenue using 
# the current mean and standard deviation, another list with 12 months 
# of revenue using the predicted mean and standard deviation, and a 
# third list combining your first two lists. You can assume a normal 
# distribution. Round each number to the nearest $1,000.


# The import command loads the functions from the random module
import random

# The randrange function returns an integer in range(a,b).
# Here we ask for a random integer n months, where 1 <= n <= 12.
random.randrange(1,13)

# We use a for loop and the randrange command to create a list of 12 integers
random_list = []
for i in range(12):
  random_list.append(random.randrange(1,13))

# output the list of random numbers
print("Random list of 12 numbers ranging from 1 to 12")
print(random_list)
print("")

# We double-check that random_list really has 12 elements in it
len(random_list)

# The sort() function will sort our list in ascending order.
random_list.sort()

# output the sorted list of 12 random numbers
print("Random list of 12 numbers ranging from 1 to 12: SORTED")
print(random_list)
print("")

# Let's compute the median of our sorted list by hand. Remember that list indexing starts at 0.
median_calculated = (random_list[5]+random_list[6])/2
print("median calculated by hand = %f" % (median_calculated))

# We use the as command to abbreviate the name of the numpy module
import numpy as np

# calculate the mean for the list of the 12 random numbers
mean_random_list = np.mean(random_list)
print("mean = %f" % (mean_random_list))

# calculate the median for the list of the 12 random numbers
median_random_list = np.median(random_list)
print("median = %f" % (median_random_list))
print("")

# for the current revenue
# We make a list of 12 values
# mean = $100000
# std = 12000
current_list = []
for i in range(12):
  current = int(random.normalvariate(100000, 12000))
  if current < 0:
    current  =  88000
  if current > 120000:
    current  = 112000
  current_list.append(current)

# output the list of 12 current revenue values
print("Current annual revenue by month is:")
print(current_list)
print("")

# for the proposed revenue
# We make a list of 12 values
# mean = $120000
# std = 25000
proposed_list = []
for i in range(12):
  proposed = int(random.normalvariate(120000, 25000))
  if proposed < 0:
    proposed  =  95000
  if proposed > 120000:
    proposed  = 145000
  proposed_list.append(proposed)

# output the list of 12 proposed revenue values
print("Proposed annual revenue by month is:")
print(proposed_list)
print("")

# round the list of 12 current revenue values to the nearest $1000
current_rnd_list = []
for r in current_list:
  current_rnd_list.append(round(r, -3))

# round the list of 12 proposed revenue values to the nearest $1000
proposed_rnd_list = []
for r1 in proposed_list:
  proposed_rnd_list.append(round(r1, -3))

# output the list of 12 current revenue rounded values
print("Current annual revenue (rounded) by month is:")
print(current_rnd_list)
print("")

# calculate the total current revenue for the year
current_annual = 0
for ca in current_rnd_list:
  current_annual += ca

# output the total current annual revenue 
print("Annual Currrent Revenue (rounded) = %d" % (current_annual))
print("")

# output the list of 12 proposed revenue rounded values
print("Proposed annual revenue (rounded) by month is:")
print(proposed_rnd_list)
print("")

# calculate the total proposed revenue for the year
proposed_annual = 0
for pa in proposed_rnd_list:
  proposed_annual += pa

# output the total proposed annual revenue 
print("Annual Proposed Revenue (rounded) = %d" % (proposed_annual))
print("")

# Create a third list by combining the propose revenue - current revenue rounded list values
combined_list = []
c = 0
while c < 12:
  proposed_rev = proposed_rnd_list[c]
  current_rev  = current_rnd_list[c]
  combined     = proposed_rev - current_rev
  combined_list.append(combined)
  c += 1

# ouput the third list of combined values
print("Difference of the Rounded Proposed Revenue minus Rounded Current Revenue:")
print(combined_list)
print("")

# calculate the change in annual revenue
revenue_change = 0
for combined_value in combined_list:
  revenue_change += combined_value

# output the change in annual revenue
print("Change in annual rounded revenue (Proposed Revenue - Current Revenue) = %d" % (revenue_change))
print("")
````

# Problem 2(a).  Bus arrival times.

Shuttle buses arrive at an airport to fetch passengers with an average interval of 15 minutes.  Their actual arrival times follow a [Poisson process](http://preshing.com/20111007/how-to-generate-random-timings-for-a-poisson-process/).  Use the ```random``` module to generate a list of 50 different bus intervals, rounded to the nearest tenth of a minute.  For example, your list might begin ```[11.2, 34.1, 18.8, 23.5, ...]```.

What is the shortest waiting time in your list?  What is the longest waiting time?

````python
# AUTHOR: Brian Pankau
# DATE: 16 June 2017
# CLASS: UWLAX DS710 SUMMER 2017 
# ASSIGNMENT: 4-2ab Python program 
# EXCUTE: python 
#
# Problem 2a: Bus Arrival Times
# Shuttle buses arrive at an airport to fetch passengers with an average 
# interval of 15 minutes. Their actual arrival times follow a Poisson 
# process. Use the random module to generate a list of 50 different bus 
# intervals, rounded to the nearest tenth of a minute. For example, your 
# list might begin [11.2, 34.1, 18.8, 23.5, ...].
# What is the shortest waiting time in your list? 
# What is the longest waiting time?


# The import command loads the functions from the random module
import random

# the average arrival rate of the bus is every 15 minutes or (15/60) 
# rate parameter = lambd = (15/60) = 0.25 buses per minute
lambd = 0.25

# generate 50 different bus intervals, rounded to the nearest tenth of a minute
bus_arrival_list = []
b = 0
while b < 50:
  bus_arrival_list.append(round(random.expovariate(lambd), 1))
  b +=1

# output bus_arrival_list
print("50 different bus intervals with a frequency of 0.25")
print(bus_arrival_list)
print("")

# Find the shortest waiting time in the bus_arrival_list
# Find the longest waiting time in the bus_arrival_list
shortest_time = bus_arrival_list[0]
longest_time = bus_arrival_list[0]
for t in bus_arrival_list:
  if (t < shortest_time):
    shortest_time = t
  if (t > longest_time):
    longest_time = t

# output the shortest waiting time in the bus_arrival_list
# output the longest waiting time in the bus_arrival_list
print ("The shortest wait time = {0:.1f}".format(shortest_time))
print ("The longest  wait time = {0:.1f}".format(longest_time))
print("")


# problem 2(b): Comulative Waiting Times
# Suppose the first bus arrives at the measured number of minutes after 
# midnight. The bus company wants to track the time each bus arrived, 
# measured in minutes after midnight. Make a list of the number of minutes 
# after midnight that each bus arrived at, using the list you generated in 
# part (a). With our example in part (a), for instance, the answer would 
# start [11.2, 45.3, 64.1, 87.6, ...].
# Using the list generated in 2(a), at what time does the 50th bus arrive? 
# Print the time in the format HH:MM AM/PM where HH is the hour and MM is 
# the minute. [Do not generate a new list; make sure to use the list of 
# intervals already generated in 2(a).]

comulative_bus_arrival_list = []
comulative_arrival_time = 0
for cb in bus_arrival_list:
  comulative_arrival_time += cb
  comulative_bus_arrival_list.append(comulative_arrival_time)

# create a list of round comulative arrival times
comulative_bus_arrival_rounded_list = []
for cbr in comulative_bus_arrival_list:
  rnd_time = round(cbr, 1)
  comulative_bus_arrival_rounded_list.append(rnd_time)

# output the comulative list
print("The comulative arrival times are:")
print(comulative_bus_arrival_rounded_list)
print("")

# output comulative time for the 50th bus arrival
print ("The 50th bus comulative arrival time in minutes = {0:.1f}".format(comulative_arrival_time))
print("")

# calculate the arrival time in HH:MM
com_hours = int(comulative_arrival_time / 60)
com_minutes = int(comulative_arrival_time % 60)
print("The 50th bus arrives at {:0>2d}:{:0>2d} AM".format(com_hours, com_minutes))
print("")
````

# Problem 3(a).  Chocolate and the Nobel.

Researchers have observed a (presumably spurious) correlation between per capita chocolate consumption and the rate of Nobel prize laureates: see [Chocolate Consumption, Cognitive Function, and Nobel Laureates](http://www.nejm.org/doi/full/10.1056/NEJMon1211064).  In this problem, we will create some sample data to simulate this relationship.

Write Python code to produce a list of 50 ordered pairs $(c,n)$, where $c$ represents chocolate consumption in kg/year/person and $n$ represents the number of Nobel laureates per 10 million population.  The values for $c$ should be random numbers (not necessarily integers!) between 0 and 15.  You may assume that $c$ and $n$ are related by

$n = 0.4\cdot c-0.8$.

However, it is not possible for a nation to have a negative number of Nobel laureates, so if your predicted value of $n$ is less than 0, replace that value by 0.

Report your values of $c$ and $n$ to 2 decimal places.  Print your list of ordered pairs.

````python
# AUTHOR: Brian Pankau
# DATE: 16 June 2017
# CLASS: UWLAX DS710 SUMMER 2017 
# ASSIGNMENT: 4-3abcd Python program 
# EXCUTE: python 
#
# Problem 3(a): Chocolate and the Nobel
#Researchers have observed a (presumably spurious) correlation between per 
# capita chocolate consumption and the rate of Nobel prize laureates: 
# see Chocolate Consumption, Cognitive Function, and Nobel Laureates. 
# In this problem, we will create some sample data to simulate this relationship.
# Write Python code to produce a list of 50 ordered pairs (c,n), where c 
# represents chocolate consumption in kg/year/person and n 
# represents the number of Nobel laureates per 10 million population. 
# The values for c should be random numbers (not necessarily integers!) 
# between 0 and 15. You may assume that c and n are related by n=0.4c−0.8.
# However, it is not possible for a nation to have a negative number of 
# Nobel laureates, so if your predicted value of n is less than 0, replace 
# that value by 0. 
# Report your values of c and n to 2 decimal places. 
# Print your list of ordered pairs.

# The import command loads the functions from the random module
import random

# generate 50 key:value pairs of c:n
n_list = []
c_list = []
i = 0
while i < 50:
  # calculate c
  c = random.uniform(0,15)
  
  # calculate n
  n = ((0.4 * c) - 0.8)
  if (n < 0):
    n = 0
  
  # append the c & n values to their lists
  n_list.append(n)
  c_list.append(c)
  
  i += 1

# create a dictionary of c:n pairs
i = 0
ordered_pair_dict = {}
#print("'i' : 'c' : 'n'") # debug
for c, n in zip(c_list, n_list):
  ordered_pair_dict[c] = n
  #print("%d : %f : %f" % (i, c, n)) # debug
  i += 1

# verify list lengths
#print("*** length of n_list = %d" % (len(n_list)))  # debug
#print("*** length of c_list = %d" % (len(c_list)))  # debug
#print("*** length of ordered_pair_dict = %d" % (len(ordered_pair_dict))) # debug

# output the ordered pair list formated to 2 decimal places
print("")
print("PROBLEM 3(A) OUTPUT: ORDERED_PAIR_LIST")
print("'i' : 'c' : 'n'")
i = 0
for c, n in ordered_pair_dict.items():
  print("{:d} : {:.2f} : {:.2f}".format(i, c, n))
  i += 1

print("")


# Problem 3(b): Error Term
# Our list of data from part (a) is not a good simulation of real-world data, 
# because it is perfectly linear. Starting with the c and n values you 
# generated in part (a), generate new ne values, using the following formula:
# ne=n+ϵ. 
# Here ϵ should be a random variable with normal distribution, mean 0, and 
# standard deviation 1. Using the list of ordered pairs generated in 3(a), 
# create a new list of 50 ordered pairs (c,n e ).
# Again, your simulated data should not predict negative numbers of Nobel 
# laureates. Again, do not generate a new list; make sure to use the list of 
# ordered pairs already generated in 3(a).
# Print your new list of ordered pairs.

# for each n in list (a), adjust the n by the error term and store in new list
# print("'i' : 'c' : 'n' : 'et' : 'ne'") # debug
error_dict = {}
i = 0
#print("'i' : 'c' : 'n' : 'et' : 'ne'") #debug
for c, n in ordered_pair_dict.items():
  # calculate the error term: normal distribution, mean=0, std=1
  et = random.normalvariate(0,1)
  
  # calculate error term (et), add error term to n 
  ne = n + et
  #print("{:d} : {:.2f} : {:.2f}: {:.2f}: {:.2f}".format(i, c, n, et, ne)) # debug
  
  # generate new error dict
  error_dict[c] = ne
  i += 1

# output new list with updated error term
print("PROBLEM 3(B) OUTPUT: ORDERED_PAIR_LIST ADUSTED WITH ERROR TERM")
print("'i' : 'c' : 'ne'")
i = 0

for c, ne in error_dict.items():
  print("{:d} : {:.2f} : {:.2f}".format(i, c, ne))
  i += 1

print("")


# Problem 3(c): Winners and Losers
# Make a new list consisting of all of the ordered pairs from your list from 
# part (b) such that ne > (0.4*c)−0.8.
# Print this new (shorter) list. 
# Use Python to calculate how many items are in your list of winners.

# create winners_dict (c:ne) by including only the ne:c pairs where ne > (0.4*c)-0.8
winners_dict = {}
i = 0
#print("'i' : 'c' : 'ne' : 'it'") # debug
for c, ne in error_dict.items():
  # calculate inclusion term (it)
  it = ((0.4 * c) - 0.8)
  
  # test inclusion term against ne
  if (ne > it):
    winners_dict[c] = ne
    #print("{:d} : {:.2f} : {:.2f} : {:.2f}".format(i, c, ne, it)) #debug
  # do not add the c:ne pair to the winners_dict
  else:
    #print("{:d} : {:.2f} : {:.2f} : {:.2f} ****".format(i, c, ne, it)) # debug
    pass
  
  i += 1

print("")

# print the winners_dict
i = 0
print("PROBLEM 3(C) OUTPUT: WINNERS_LIST")
print("'i' : 'c' : 'ne'")
for c, ne in winners_dict.items():
  print("{:d} : {:.2f} : {:.2f}".format(i, c, ne))
  i += 1

print("")

# output the number of items in the winners_dict
print("The winners dictionary has %d items" % (i))
print("")


# Problem 3(d): Preparing for R
# For future use, split your list of ordered pairs from Problem 3(b) (not 3(c)!) 
# into two lists. The first list should contain the c values, and the second list 
# should contain the ne values.

# create new list with updated error term
new_c_list = []
new_ne_list = []
# k = ne
# v = c
for c, ne in error_dict.items():
  new_c_list.append(c)
  new_ne_list.append(ne)

print("")

# output new lists
print("PROBLEM 4(D) OUTPUT: c & ne lists")
print("C LIST:")
print(new_c_list)
print("")
print("NE LIST:")
print(new_ne_list)
print("")
````

````python
# AUTHOR: Brian Pankau
# DATE: 16 June 2017
# CLASS: UWLAX DS710 SUMMER 2017 
# ASSIGNMENT: 4-3abcd Python program 
# EXCUTE: python 
#
# Problem 3(a): Chocolate and the Nobel
#Researchers have observed a (presumably spurious) correlation between per 
# capita chocolate consumption and the rate of Nobel prize laureates: 
# see Chocolate Consumption, Cognitive Function, and Nobel Laureates. 
# In this problem, we will create some sample data to simulate this relationship.
# Write Python code to produce a list of 50 ordered pairs (c,n), where c 
# represents chocolate consumption in kg/year/person and n 
# represents the number of Nobel laureates per 10 million population. 
# The values for c should be random numbers (not necessarily integers!) 
# between 0 and 15. You may assume that c and n are related by n=0.4c−0.8.
# However, it is not possible for a nation to have a negative number of 
# Nobel laureates, so if your predicted value of n is less than 0, replace 
# that value by 0. 
# Report your values of c and n to 2 decimal places. 
# Print your list of ordered pairs.

# The import command loads the functions from the random module
import random

# generate 50 key:value pairs of c:n
n_list = []
c_list = []
i = 0
while i < 50:
  # calculate c
  c = random.uniform(0,15)
  
  # calculate n
  n = ((0.4 * c) - 0.8)
  if (n < 0):
    n = 0
  
  # append the c & n values to their lists
  n_list.append(n)
  c_list.append(c)
  
  i += 1

# create a dictionary of c:n pairs
i = 0
ordered_pair_dict = {}
#print("'i' : 'c' : 'n'") # debug
for c, n in zip(c_list, n_list):
  ordered_pair_dict[c] = n
  #print("%d : %f : %f" % (i, c, n)) # debug
  i += 1

# verify list lengths
#print("*** length of n_list = %d" % (len(n_list)))  # debug
#print("*** length of c_list = %d" % (len(c_list)))  # debug
#print("*** length of ordered_pair_dict = %d" % (len(ordered_pair_dict))) # debug

# output the ordered pair list formated to 2 decimal places
print("")
print("PROBLEM 3(A) OUTPUT: ORDERED_PAIR_LIST")
print("'i' : 'c' : 'n'")
i = 0
for c, n in ordered_pair_dict.items():
  print("{:d} : {:.2f} : {:.2f}".format(i, c, n))
  i += 1

print("")


# Problem 3(b): Error Term
# Our list of data from part (a) is not a good simulation of real-world data, 
# because it is perfectly linear. Starting with the c and n values you 
# generated in part (a), generate new ne values, using the following formula:
# ne=n+ϵ. 
# Here ϵ should be a random variable with normal distribution, mean 0, and 
# standard deviation 1. Using the list of ordered pairs generated in 3(a), 
# create a new list of 50 ordered pairs (c,n e ).
# Again, your simulated data should not predict negative numbers of Nobel 
# laureates. Again, do not generate a new list; make sure to use the list of 
# ordered pairs already generated in 3(a).
# Print your new list of ordered pairs.

# for each n in list (a), adjust the n by the error term and store in new list
# print("'i' : 'c' : 'n' : 'et' : 'ne'") # debug
error_dict = {}
i = 0
#print("'i' : 'c' : 'n' : 'et' : 'ne'") #debug
for c, n in ordered_pair_dict.items():
  # calculate the error term: normal distribution, mean=0, std=1
  et = random.normalvariate(0,1)
  
  # calculate error term (et), add error term to n 
  ne = n + et
  #print("{:d} : {:.2f} : {:.2f}: {:.2f}: {:.2f}".format(i, c, n, et, ne)) # debug
  
  # generate new error dict
  error_dict[c] = ne
  i += 1

# output new list with updated error term
print("PROBLEM 3(B) OUTPUT: ORDERED_PAIR_LIST ADUSTED WITH ERROR TERM")
print("'i' : 'c' : 'ne'")
i = 0

for c, ne in error_dict.items():
  print("{:d} : {:.2f} : {:.2f}".format(i, c, ne))
  i += 1

print("")


# Problem 3(c): Winners and Losers
# Make a new list consisting of all of the ordered pairs from your list from 
# part (b) such that ne > (0.4*c)−0.8.
# Print this new (shorter) list. 
# Use Python to calculate how many items are in your list of winners.

# create winners_dict (c:ne) by including only the ne:c pairs where ne > (0.4*c)-0.8
winners_dict = {}
i = 0
#print("'i' : 'c' : 'ne' : 'it'") # debug
for c, ne in error_dict.items():
  # calculate inclusion term (it)
  it = ((0.4 * c) - 0.8)
  
  # test inclusion term against ne
  if (ne > it):
    winners_dict[c] = ne
    #print("{:d} : {:.2f} : {:.2f} : {:.2f}".format(i, c, ne, it)) #debug
  # do not add the c:ne pair to the winners_dict
  else:
    #print("{:d} : {:.2f} : {:.2f} : {:.2f} ****".format(i, c, ne, it)) # debug
    pass
  
  i += 1

print("")

# print the winners_dict
i = 0
print("PROBLEM 3(C) OUTPUT: WINNERS_LIST")
print("'i' : 'c' : 'ne'")
for c, ne in winners_dict.items():
  print("{:d} : {:.2f} : {:.2f}".format(i, c, ne))
  i += 1

print("")

# output the number of items in the winners_dict
print("The winners dictionary has %d items" % (i))
print("")


# Problem 3(d): Preparing for R
# For future use, split your list of ordered pairs from Problem 3(b) (not 3(c)!) 
# into two lists. The first list should contain the c values, and the second list 
# should contain the ne values.

# create new list with updated error term
new_c_list = []
new_ne_list = []
# k = ne
# v = c
for c, ne in error_dict.items():
  new_c_list.append(c)
  new_ne_list.append(ne)

print("")

# output new lists
print("PROBLEM 4(D) OUTPUT: c & ne lists")
print("C LIST:")
print(new_c_list)
print("")
print("NE LIST:")
print(new_ne_list)
print("")
````
