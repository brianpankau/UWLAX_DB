# Problem 1(a). Pig Latin.

In "Pig Latin", English words are transformed according to the following rules:

* If the word begins with one or more consonants, those consonants are moved to the end of the word, followed by "ay":

  * pig -> igpay
  * Latin -> Atinlay
  * trash -> ashtray

* If the word begins with a vowel, it is followed by "yay":

  * apple -> appleyay
  * out -> outyay.

* If the word begins with "qu", both of these letters are moved to the end of the word, followed by "ay":

  * quarter -> arterquay
  
Note that more than one consonant may be moved to the end of the word. Note also that "y" functions as a consonant at the start of words. Note also that capitalization should be preserved after manipulating the word. 

Write Python code that takes a word and converts it to Pig Latin.  If the input consists of multiple words or contains punctuation, your code should print a suitable error message. 

````python
# AUTHOR: Brian Pankau
# DATE: 16 June 2017
# CLASS: UWLAX DS710 SUMMER 2017 
# ASSIGNMENT: 5 Python program 
# EXCUTE: python 
#
# Problem 1(a) : 
#    Write Python code that takes a word and converts it to Pig Latin. 
#    01) If the word begins with one or more consonants, 
#        those consonants are moved to the end of the word, followed by "ay".
#    02) If the word begins with a vowel, it is followed by "yay".
#    03) If the word begins with "qu", both of these letters are moved to 
#        the end of the word, followed by "ay".
#    04) If the input consists of multiple words or contains punctuation, 
#        your code should print a suitable error message.
#    05) Note that more than one consonant may be moved to the end of the word. 
#    06) Note also that "y" functions as a consonant at the start of words. 
#    07) Note also that capitalization should be preserved after manipulating the word. 
#
# INPUT TEST SCENARIOS
# T1) Input contains 2+ words separated by space(s): "one two" => ERROR
# T2) Input contains a single letter article 'a': "a" => "ayay", "A" => "Ayay"
# T3) Input contains a single letter article not in ('a', 'A')
# T4) Input word contains punctuation marks => ERROR
# T5) Input word is all consonants => ERROR
# T6) Input is a word beginning with a vowel => append a "yay" onto the word
# T7) Input is a word beginning with "qu" or "Qu" => move the letters to the end & append "ay"
# T8) Input is a word beginning with one or more constants => move constants to end & append "ay"
#

# set input
test_list1a = ["one Two","a","A","s","T","no!","tst","apple","Apple","quater","Quater","pig","Pig","latin","Latin","trash","Trash"]
test_list1b = ["orange", "yellow", "Strip", "quarter", "schmooze", "a", "Pig Latin", "Ke$ha"]
test_list   = test_list1a
print("The test input list = '"'%s'"'" % (test_list))
print("")
print("PROGRAM OUTPUT")

# define constants punctuation marks
punct_list = ['!', '"', "'", '(', ')', ',', '-', '.', '...', ':', ';', '?', '[', ']', '{', '}', '$']
  
# define vowels
vowels = ['A', 'E', 'I', 'O', 'U', 'a', 'e', 'i', 'o', 'u']
  
for word in test_list:
  word_list = list(word)
  
  # set processing of word completion flag
  done = False
  
  # validate the word
  valid = True
  for i, v in enumerate(word_list,0):
    # handle test case 1: multiple words
    if v == ' ':
      print("*** ERROR: The word '"'%s'"' has a space indicating that there are multiple words" % (word))
      valid = False
      done = True
    
    # handle test case 2: single letter article "a" or "A"
    elif (len(word) == 1):
      if ((word[0] == 'a') | (word[0] == 'A')):
        pig_word = word
        pig_word += "yay"
        # if the word was capitalized, then capitalize the translated word
        if (word.istitle()):
          pig_word = pig_word.title()
        
        print ("==> word = '"'%s'"', pig latin translation = '"'%s'"'" % (word, pig_word))
        done = True
      
      # handle test case 3: not the single letter article 'a' or 'A'
      else:
        print("*** ERRROR: The word '"'%s'"' is a single letter but is not the article set ('a' | 'A')" % (word))
        valid = False
        done = True
    
    # handle test case 4: check that no letters in the word are within the punctuation mark list    
    else:
      for j, p in enumerate(punct_list, 0):
        if(v == p):
          print("*** ERROR: Punctuation mark '"'%s'"' found in word '"'%s'"' at letter index '"'%s'"'" % (p, word, i))
          valid = False
          done = True
          break
  
  # Test case 5: Verify that the word is not all consonants
  myfound = False
  if (not done):
    for k in word_list:
      if k in vowels:
        myfound = True
        break
    if (myfound == False):
      valid = False
      done = True
      print ("*** ERROR: No vowels found in word '"'%s'"'" % (word))
  
  # word is validated, translate the word
  if (not done):
    # handle test case 6: the first letter begins with a vowel
    if (word[0] in vowels):
      pig_word  = word
      pig_word += "yay"
      # if the word was capitalized, then capitalize the translated word
      if (word.istitle()):
        pig_word = pig_word.title()
      
      print("==> The word '"'%s'"' starts with a vowel, pig latin translation = '"'%s'"'" % (word, pig_word))
      done = True
    
    # handle test case 7: the first & second letter are 'qu' or "Qu"
    elif (((word[0] == 'Q') | (word[0] == 'q')) & ((word[1] == 'U') | (word[1] == 'u'))):
      pig_word  = word[2:]
      pig_word += word[0:2]
      pig_word += "ay"
      # if the word was capitalized, then capitalize the translated word
      if (word.istitle()):
        pig_word = pig_word.title()
      
      print("==> The word '"'%s'"' starts with 'QU/Qu/qU/qu', pig latin translation = '"'%s'"'" % (word, pig_word))
      done = True
    
    else:
      # handle test case 8: constants, find the first vowel
      first_vowel = 0
      while first_vowel <= len(word): 
        if word[first_vowel] not in vowels:
          first_vowel += 1
        else:
          break
      
      pig_word  = word[first_vowel:]
      pig_word += word[0:first_vowel]
      pig_word += "ay"
      # if the word was capitalized, then capitalize the translated word
      if (word.istitle()):
        pig_word = pig_word.title()
      
      print("==> The word '"'%s'"' starts with consonant, pig latin translation = '"'%s'"'" % (word, pig_word))
      done = True

````

# Problem 2(a).  Letter Frequencies.

The files ```encryptedA.txt``` and ```encryptedB.txt``` contain two different encrypted messages on similar topics.  One message was originally in English and one was in Welsh.  

Write Python code that reads a text file into memory and creates a dictionary object with a frequency count for each letter.  For example, for ```encryptedA.txt```, your output should contain the key:value pairs ```'a': 78``` and ```'b': 31 ```.  Do not distinguish between uppercase and lowercase letters, and ignore punctuation. If a given letter does not appear in the text, there should be a key:value pair with value 0. 

Use Python to determine which letter has the highest frequency in each text file, and print the result.

````python
# AUTHOR: Brian Pankau
# DATE: 16 June 2017
# CLASS: UWLAX DS710 SUMMER 2017 
# ASSIGNMENT: 5-2ab Python program 
# EXCUTE: python 
#
# Problem 2(a) Letter Frequences: 
#
# The files encryptedA.txt and encryptedB.txt contain two different encrypted 
# messages on similar topics. One message was originally in English and one 
# was in Welsh. 
# Write Python code that reads a text file into memory and creates a dictionary 
# object with a frequency count for each letter. For example, for 
# encryptedA.txt, your output should contain the key:value pairs 
# 'a': 78 and 'b': 31. Do not distinguish between uppercase and 
# lowercase letters, and ignore punctuation. 
# If a given letter does not appear in the text, there should be 
# a key:value pair with value 0. 
# Use Python to determine which letter has the highest frequency 
# in each text file, and print the result.

# Problem 2(b): Formatting for R
#
# Write your dictionaries with frequency counts to .csv files, with one column 
# for the key and one column for the frequency counted.


# set the input file list for the file names to be processed
filename_list = ['C:/Temp1/encryptedA.txt', 'C:/Temp1/encryptedB.txt']

# process each filename in filename_list
for fn in filename_list:
  # store the filename into a variable
  filename = fn

  # print the filename
  print("PROBLEM 2A: LETTER FREQUENCES OUTPUT")
  print("FILENAME = %s" % (filename))
  
  # open, read, and close the contents of the file into a variable
  with open(filename) as f_in:
    filedata = f_in.readlines()
    
  # create a list of keys from the unique characters in the input file
  for n, line in enumerate(filedata, 1):
    i = 0
    while (i < len(line)):
      i += 1
  char_list = list(line)
  key_list = set(char_list)          # list of unique characters in input list
  value_list = [0]*len(key_list)     # list of empty values matching one-to-one to the key_list
  
  # create a dictionary (char_dict) of key:value pairs using key_list:value_list
  char_dict = {}
  for key, value in zip(key_list, value_list):
    char_dict[key] = value
  
  # count the number of occurances in the input and update each key:value in char_dict
  for letter in char_list:
    value = char_dict.get(letter, 1)
    char_dict[letter] = value + 1
  
  # create a dictionary (alpha_dict) of alphabetical characters with empty values
  alpha_key   = ['a','b','c','d','e','f','g','h','i','j','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
  alpha_value = [0]*len(alpha_key)
  alpha_dict  = {}
  for key, value in zip(alpha_key, alpha_value):
    alpha_dict[key] = value
  
  # copy the values from the char_dict to the alpha_dict for each key
  # track which key has the highest value
  high_key   = 'a'
  high_value = 0
  for alpha in alpha_key:
    value = char_dict.get(alpha, 0)
    alpha_dict[alpha] = value
    if(high_value < value):
      high_value = value
      high_key = alpha
    print(alpha, value)
  print("High key = %s, High value = %d" % (high_key, high_value))
  print("")
  
  
  # create a .csv file with line as a key and a value separated by a comma
  print("PROBLEM 2B: FORMATTED OUPUT FOR R")
  cvs_filename = filename+".csv"
  print("FILENAME = %s" % (cvs_filename))
  
  # open the ouput file in write mode
  fn_out = open(cvs_filename, 'w')
  
  # convert dictionary to raw
  for alpha in alpha_key:
    value = char_dict.get(alpha, 0)
    fn_out.write("%s,%d\n" % (alpha, value))
  
  # close the ouptut file
  fn_out.close()
  
  # add output separators for each processed input file 
  print("")
  print("------------------------------------------------")
  print("")
````
